```python
orig = cv2.imread('content.png')
style = cv2.imread('style_0.png')
result = cv2.imread('result.png')
```

```python
%matplotlib inline
# recombine the color channels
def order_colors(img):
    channels = blue, green, red = np.rollaxis(img, 2)
    rgb_img = np.dstack([red, green, blue])
    return rgb_img

plt.imshow(order_colors(orig )); plt.axis("off"); plt.title("Original");
plt.imshow(order_colors(style )); plt.axis("off"); plt.title("Original");
plt.imshow(order_colors(result  )); plt.axis("off"); plt.title("Original");
```

```python
plt.figure(figsize = (6,6))
plt.imshow(order_colors(orig )); plt.axis("off"); plt.title("Original");
```
